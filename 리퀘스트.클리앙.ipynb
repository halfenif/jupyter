```python
import re
import requests
from bs4 import BeautifulSoup


url = 'https://www.clien.net/service/board/park/10719569'

out = requests.get(url).text
title = re.findall('<title>(.*)</title>', out)[0]

lxml = BeautifulSoup(out,'lxml')
time = lxml.find('div', attrs={"class": "post-time"}).text.strip()
body = lxml.find('div', attrs={"class": "post-article fr-view"}).text
user = lxml.find('button', attrs={"class": "dropdown-toggle nick"}).text.strip()


print(user)
print(title)
print(time)
print('------------------------------')
print(body)
```

```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException


url ="http://m.imdb.com/feature/bornondate"
#Driver
driver = webdriver.Firefox(executable_path="C:/Users/junye/geckodriver-v0.16.1-win64/geckodriver.exe")

#Get
driver.get(url)


try:
    posterslist = driver.find_elements_by_class_name("poster")
    #print(posterslist)
    
    for item in posterslist:
        title = item.find_element_by_class_name("title").text
        detail = item.find_element_by_class_name("detail").text
        
        print(title + ' : ' + detail)
except NoSuchElementException:
    print('NoSuchElementException')

#End
driver.quit()
print('----------------End]')
```
